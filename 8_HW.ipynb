```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def f1(params, x):
    return params[0]/(x - params[1]) + params[2]


x = np.linspace(0.01, 2, 100)
p = [0.36, 0, 1]
y = f1(p, x)

plt.plot(x, y)
```

```python
def NormalMatr(params, x):
    #x[params[1] - x < 1e19] = 1e19
    x = x.reshape(100, 1)
    return np.concatenate((2 * params[0] / x, np.ones(shape = (100, 1))), axis = 1) 

def pred(params, x):
    return params[0] ** 2 / x + params[1]
```

```python
Lg = LinearRegression(fit_intercept = False)
old_params = [1e4, 100]
for i in range(20):
    ri = y - pred(old_params, x)
    Lg.fit(NormalMatr(old_params, x), ri)
    old_params += Lg.coef_
    print(old_params)
    
```

```python
def f(w, x):
    return w[0] * x ** w[1]+ w[2] * x ** w[3]

def df_dx(w1, w2, w3, w4, x):
    return w1 * w2 * x ** (w2-1) + w3 * w4 * x ** (w4-1)

def grad(w, x, y):
    r = f(w, x) - y
    #x = x.reshape(len(x), 1)
    a1 = (x ** w[1] * r).mean()
    a2 = ((w[0] *  x **w[1] * np.log(x)) * r).mean()
    a3 = (x ** w[3] * r).mean()
    a4 = ((w[2] * x ** w[3] * np.log(x))* r).mean()
    return np.array([a1, a2, a3, a4])
```
