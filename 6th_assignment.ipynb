**Задание 1**

Пусть дана выборка точек на прямой ${x_i}$. Максимизируйте правдоподобие (или его логарифм) в гауссовой
вероятностной модели:
$$\prod\limits_{i}p(x_i) → \max\limits_{{μ, σ}}\\ p(x) = \frac{1}{\sqrt{2πσ^2}}e^{-\frac{{(x-μ)}^2}{2σ^2}}$$


Решение:

$$P = \prod\limits_{i=1}^n\frac{1}{\sqrt{2 \pi \sigma^2}}\cdot \exp{\left(-\frac{{(x_i - \mu)}^2}{2 \sigma^2} \right)}$$,

$$logP = \sum\limits_{i=1} ^n \left(\frac{1}{2} log(2 \pi \sigma^2) -\frac{{(x_i - \mu)}^2}{2 \sigma^2}\right)\$$

 $$\frac{\partial logP}{\partial \sigma^2} = -\frac{n}{\sigma^2} + \frac{1}{2 \sigma^4} \sum\limits_{i=1}^n {(x_i - \mu)}^2 = 0$$

 $$\frac{\partial logP}{\partial \mu} = \sum\limits_{i=1}^n \frac{(x_i - \mu)}{2\sigma^2} = 0 ⇒ \boxed{μ = \sum\limits_{i=1}^n \frac{x_i}{n}} - ср. арифм$$
 $$ ⇒ \boxed{\sigma^2 = \frac{1}{n} \sum\limits_{i=1}^n {(x_i - \mu)}^2}$$

**Задание 3**

Ультрачувствительный тест от короновируса ошибается в $1\%$ случаев (как в одну, так и в другую сторону). В
данный момент в популяции доля заболевших $10^{−5}$. Петя получил положительный тест на коронавирус. С какой вероятностью он действительно болеет коронавирусом?

$ p(б) = 10^{-5} ⇒ p(здор.) = 1 - p(\delta) = 0.99999 $

$ p(+|б.) = 0.99 $

$ p(+|здор.) = 0.01 $

$p(+) = p(б)\cdot p(+|б) + p(здор.)\cdot p(+|здор.) = 10^{-5}\cdot 0.99+0.99999 \cdot 0.01 = 0.0100098 $

Ответ: $\boxed{p(б|+) =\frac{p(+|б) p(б)}{p(+)} \approx 0.00099} $

**Задача 6**

Для стандартного набора данных для задачи регрессии (см. например load_diabetes из sklearn.datasets) продемонстрируйте, как веса обращаются в ноль по мере увеличения коэффицинета L1-регуляризации. Разрешается использовать библиотечную реализацию регрессии.

```python
from sklearn.datasets import load_diabetes
```

```python
Data_X, data_y = load_diabetes(return_X_y = True, as_frame = True)
Data_X.shape

alphas = np.linspace(0.1, 1e1, 100)
coefs = np.empty(shape = (10, 1))
model = Lasso(alpha = 0.1, normalize = True)

model.fit(Data_X, data_y)
```

```python
model.coef_
```

```python
for a in alphas: 
    model = Lasso(alpha = a, normalize = True)
    model.fit(Data_X, data_y)
    coefs = np.concatenate((coefs, np.array(model.coef_.reshape(10, 1))), axis = 1)
```

```python
for i in range(10):
    plt.plot(np.arange(101), coefs[i, :])
```
